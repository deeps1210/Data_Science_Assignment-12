```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import sqlite3
import sqlalchemy
from pandasql import sqldf
```

```python
pysqldf = lambda q: sqldf(q, globals())
```

```python
sqladb=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',
                   names=['Age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship',
                          'race','sex','capital-gain','capital-loss','hours-per-week','native-country','>50K, <=50K'])
```

```python
#Trim all string(object) columns for blank spaces
def trimallcolumns(df):
    trimStrings = lambda x:x.strip() if type(x) is str else x
    return df.applymap(trimStrings)
```

```python
sqladb = trimallcolumns(sqladb)
```

```python
#replace column names "-" with "_"
sqladb.columns = sqladb.columns.str.replace('-','_')
```

```python
#5 What is the average, minimum and maximum age group for people working in different sectors
```

```python
t = """SELECT avg(Age) as Avg_Age, max(Age) as Max_Age,
            min(Age) as Min_Age, workclass
        from sqladb GROUP BY workclass;""" 
```

```python
#6. Calculate age distribution by country
```

```python
a = """SELECT case 
                when Age <18 then 'Under 18'
                when Age between 18 and 30 then '18-30'
                when Age between 31 and 60 then '31-60'
                when Age >60 then 'Above 60'
         END as age_range,
            Count(*) as count, native_country
            from sqladb
            group by age_range, native_country
            order by age_range;"""
```

```python
pysqldf(a)
```
